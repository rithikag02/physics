```python
import numpy as np
from sympy.physics.quantum.cg import CG
import pandas as pd
```

```python
def Calc(j1, m1, j2, m2, j3, m3):
    cg = CG(j1, m1, j2, m2, j3, m3)
    return cg.doit()

def lister(a, b):
    if b < a:
        a, b = b, a
    if b >= a:    
        ls = [a]
        while b != a:
            ls.append(b)
            b -= 1
        ls.sort()
        return ls

def jvals(j1, j2):
    return lister(np.abs(j1 - j2), np.abs(j1 + j2))

def mvals(j):
    ms, js = [], []
    for i in j:
        if i == 0:
            ms.extend([0])
            js.extend([0])
        else:
            ms.extend(lister(-i, i))
            js.extend([i] * int(2*i + 1))
    l = len(js)
    return np.array([js, ms]).T

def calculate_m(j1, j2):
    m1, m2 = mvals([j1]), mvals([j2])
    m = mvals(jvals(j1, j2))
    return m, m1, m2

def calculate_coef(m, m1, m2):
    list = []
    for k in m:
        for i in m1:
            for j in m2:
                if (i[0] != j[0] or (i[0] == j[0] and j[1] != i[1])):
                    cgg = Calc(i[0], i[1], j[0], j[1], k[0], k[1])
                    if cgg != 0:
                        list.append([i[0], j[0], i[1], j[1], k[0], k[1], cgg])
    return list

def print_table(list):
    return pd.DataFrame(np.array(list), columns=['j1','j2','m1','m2','j','m','Coefficient'])

def CG_coefficients(j1, j2):
    m, m1, m2 = calculate_m(j1, j2)
    list = calculate_coef(m, m1, m2)
    df = print_table(list)
    return df
```
